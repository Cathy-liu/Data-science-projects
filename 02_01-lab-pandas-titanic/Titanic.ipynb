## Step 5: Plotting
Using Matplotlib and Seaborn, create several charts showing the survival rates of different groups of people. It's fine if a handful of charts are basic (Gender, Age, etc), but what we're really looking for is something beneath the surface.


```python
from math import nan, isnan
```

```python
df_gender=data.groupby(['Sex','Survived']).count()/data.groupby(['Sex']).count()
```

```python
rate_by_gender=[]
for i in data['Sex'].unique():
    try:
        try:
            rate_by_gender.append(df_gender['Cabin'][i][1])
        except:
            rate_by_gender.append(1-df_gender['Cabin'][i][0])
    except:
        pass 
print(rate_by_gender)
```

```python
fig = plt.figure()
ax1 = fig.add_axes((0.1, 0.2, 0.8, 0.7))
ax1.set_title("Survival rate for each gender")
ax1.set_xlabel('Gender')
ax1.set_ylabel('Survival rate')
plt.bar(x=data['Sex'].unique(),height=rate_by_gender)
```

```python
#Below is to filter out those rows with missing Age values, and calculate survival rate with grouping by age.
df_age=data[data['Age'].isna()==0].groupby(['Age','Survived']).count()/data[data['Age'].isna()==0].groupby(['Age']).count()
```

```python
rate_by_age=[]

for i in data['Age'].unique():
    try:
        try:
            rate_by_age.append(df_age['Cabin'][i][1])
        except:
            rate_by_age.append(1-df_age['Cabin'][i][0])
    except:
        print(f'"{i}" values in the "Age" columne of data will be skipped.') 
        
print(rate_by_age)
```

```python
#To make a list of unique values of ages, and filter out those 'nan' values.
age_list=data['Age'].unique().tolist()
updated_age_list = [x for x in age_list if isnan(x) == False]
```

```python
#Sort the age_list first before get the population distribution 
updated_age_list.sort()
population_by_age_list=[len(data[data['Age']==i]) for i in updated_age_list]
```

```python
fig,ax1=plt.subplots(figsize=(15, 8))

#Draw bar chart for rate of survival for each age
ax1.set_title("Survival rate for each age")
ax1.set_xlabel('Age')
ax1.set_ylabel('Survival rate')
plt.bar(x=updated_age_list,height=rate_by_age,color='b')

#Draw line plot for population distribution for each age
ax2=ax1.twinx()
#ax2.set_title("Population by age")
ax2.set_ylabel("Population per age")
plt.plot(updated_age_list,population_by_age_list,color='r')
```
